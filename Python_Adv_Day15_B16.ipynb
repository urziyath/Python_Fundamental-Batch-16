```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [2,6,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title('Sales Information')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.show()
```

```python
#now adding Z axis

z = [10,5,2]
plt.title('Sales Information')
plt.xlabel('Sales')
plt.ylabel('Revenues')
plt.plot(x,y,color = 'Blue')
plt.plot(x,z,color = 'Yellow')
plt.plot(x,y, label = 'X & Y Axis')
plt.plot(x,z, label = 'X & Z Axis')
plt.legend()
plt.show()
```
